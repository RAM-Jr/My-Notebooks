```python
# Get the percentage of missing values for each attribute with missing values
def get_missing_ratio(dataset):
  dataset = dataset.isnull().sum()/len(dataset)*100 
  ratios = dataset.drop(dataset[dataset == 0].index).sort_values(ascending=False)
  return pd.DataFrame({'Missing Ratio' :ratios})
  
  
missing_data = get_missing_ratio(raw_data)
missing_data
```

```python
fig, ax = plt.subplots(figsize=(8,6))
plt.xticks(rotation='90')
sns.barplot(x=missing_data.index, y=missing_data['Missing Ratio'])
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of Missing Values', fontsize=15)
plt.title('Attributes wiht Missing Values', fontsize=15)
```

For the features with missing values of less than 5% samples, I'll simply drop the samples.
And for those with more than 5% and less than 60% I'll fill predict the missing values using a machine learning model

#### Method to deal with missing values

```python
def deal_with_missing_values(dataset,features,max_missing_ratio):
  
  if max_missing_ratio<=5:
    # for features with missing ratio <= 5%, remove the samples with missing values
    for col in features:
      dataset = dataset[dataset[col].notna()]
  elif max_missing_ratio>=60:
    # for features with missing ratio > 60%, remove the column with missing values
    dataset = dataset.drop(features,axis=1)

  else:
    # for features with missing ratio <60% and >5%, replace the missing values with the mode or mean
    for col in features:

      if dataset[col].dtypes == 'O':
        # if the feature data type is categorical, replace the missing values with the mode
        dataset[col] = dataset[col].replace(np.NaN, dataset[col].mode()[0])
      else:
        # if the feature data type is numerical, replace the missing values with the mean
        dataset[col] = dataset[col].replace(np.NaN, dataset[col].mean())

  return dataset
```

### How to deal with the missing data? 
* For features with missing ratio <= 5%, remove the samples with missing values
* For features with missing ratio > 60%, remove the column with missing values
* For features with missing ratio <60% and >5%, replace the missing values with the mode (if categorical) or mean (if numerical)

```python
features_remove_row = missing_data[missing_data['Missing Ratio']<=5].index 
features_remove_col = missing_data[missing_data['Missing Ratio']>=60].index
features_predict_missing= missing_data[missing_data['Missing Ratio']<60][missing_data['Missing Ratio']>5].index
```

```python
raw_data = deal_with_missing_values(raw_data,features_remove_row,5)
raw_data = deal_with_missing_values(raw_data,features_remove_col,60)
raw_data = deal_with_missing_values(raw_data,features_predict_missing,10)
```

```python
def find_outliers_IQR(df):
   q1=df.quantile(0.25)
   q3=df.quantile(0.75)
   IQR=q3-q1
   non_outliers = df[~((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
   return non_outliers
```

```python
d = data_no_mv.copy()
outliers = find_outliers_IQR(data_no_mv['SalePrice'])
outliers.index
```

# Skeweness

```python
numerical_attributes = data_no_mv_out.dtypes[data_no_mv_out.dtypes != "object"].index
data_no_mv_out[numerical_attributes].shape
```

There are 38 features, so I will not plot the pdf for each of these features, I'll simply compute the skeweness for each feature and then remove samples from the data based on the obtained skeweness.

```python
skewed_attributes = data_no_mv_out[numerical_attributes].skew() # after creating dummies all features became numerical

skewness = pd.DataFrame({"Variable":skewed_attributes.index, "Skewness":skewed_attributes.values})
skewness = skewness.sort_values('Skewness', ascending=[0])

fig, ax = plt.subplots(figsize=(8,6))
plt.xticks(rotation='90')
sns.barplot(x=skewness['Variable'], y=skewness['Skewness'])
plt.ylim(0,25)
plt.xlabel('Numerical Variables', fontsize=15)
plt.ylabel('Skewness', fontsize=15)
plt.title('', fontsize=15)
plt.show() 
```

```python
skewness
```

From both, the plot and the dataframe, It can be seen that many features are highly skewed. Some of them are skewed to the left (the ones with a negative skeweness value) and others are skewed to the right (those with positive a skeweness value)

```python
# Select the attributes with the skewness above the treshold, .75
skewed_attributes_right = skewed_attributes[skewed_attributes > 0.75]
skewed_attributes_left = skewed_attributes[skewed_attributes < -0.75]

# apply logaritmic transformation for the features to reduce the skewness
data_no_mv_out[skewed_attributes_right.index] = np.log(data_no_mv_out[skewed_attributes_right.index]+0.001)
data_no_mv_out[skewed_attributes_left.index] = np.log1p(data_no_mv_out[skewed_attributes_left.index]+0.001)

skewed_attributes = data_no_mv_out[numerical_attributes].skew()
skewness_new = pd.DataFrame({"Variable":skewed_attributes.index, "Skewness":skewed_attributes.values})
skewness_new = skewness_new.sort_values('Skewness', ascending=[0])

fig, ax = plt.subplots(figsize=(8,6))
plt.xticks(rotation='90')
sns.barplot(x=skewness_new['Variable'], y=skewness_new['Skewness'])
plt.ylim(0,25)
plt.xlabel('Numerical Variables', fontsize=15)
plt.ylabel('Skewness', fontsize=15)
plt.title('', fontsize=15)
plt.show() # display the plot of the skewness of the attributes after the log transformation
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        self.copy = copy
        self.with_mean = with_mean
        self.with_std = with_std
        self.columns = columns
        self.mean = None
        self.var_ = None
    
    def fit(self, X, y=None):
        self.scaler = StandardScaler(copy=self.copy,with_mean=self.with_mean,with_std=self.with_std)
        self.scaler.fit(X[self.columns],y)
        self.mean = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled,X_scaled], axis=1)[init_col_order]
```

I'll not normalize the features with only 0s and 1s because they are already in a standard form. If I scale these features I'll add noise because when I scale the data, the values that were suposed to be 0, may become another value different from 0 which will change the meaning of the value 0 itself for the feature.

```python
features_to_ommit = []

for feature in all_features:
  if len(inputs_unscaled[feature].unique()) == 2:
    features_to_ommit.append(feature)

inputs_unscaled[features_to_ommit].head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

features_to_scale = [feature for feature in inputs_unscaled.columns.values if feature not in features_to_ommit ]
features_to_scale
```

```python
scaler = CustomScaler(features_to_scale)
scaler.fit(inputs_unscaled)
```

```python
inputs_scaled = scaler.transform(inputs_unscaled)
inputs_scaled.head(2)
```
