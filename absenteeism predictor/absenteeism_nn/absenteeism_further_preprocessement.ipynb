# load the data

```python
data_preprocessed = pd.read_csv('../Absenteeism_data_preprocessed.csv')
data_preprocessed
```

```python
data_preprocessed = data_preprocessed.drop(['Unnamed: 0'],axis=1)
data_preprocessed
```

# separete the data into inputs and targets

### create the targets

```python
# for the feature 'Absenteeism Time in Hours' give value 1 for values greater than the median.
targets = np.where(data_preprocessed['Absenteeism Time in Hours']>data_preprocessed['Absenteeism Time in Hours'].median(),1,0)
targets
```

### create the inputs

```python
unscaled_inputs = data_preprocessed.drop(['Absenteeism Time in Hours',],axis=1)
# unscaled_inputs = data_preprocessed.drop(['Absenteeism Time in Hours','Distance to Work',
#        'Daily Work Load Average',
#        'Day of the Week'],axis=1)
unscaled_inputs
```

### standardize the inputs

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        self.copy = copy
        self.with_mean = with_mean
        self.with_std = with_std
        self.columns = columns
        self.mean = None
        self.var_ = None
    
    def fit(self, X, y=None):
        self.scaler = StandardScaler(copy=self.copy,with_mean=self.with_mean,with_std=self.with_std)
        self.scaler.fit(X[self.columns],y)
        self.mean = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled,X_scaled], axis=1)[init_col_order]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

features_to_ommit = ['Reason type 1', 'Reason type 2', 'Reason type 3'
                     , 'Reason type 4','Education']
features_to_scale = [feature for feature in unscaled_inputs.columns.values if feature not in features_to_ommit ]
features_to_scale
```

```python
scaler = CustomScaler(features_to_scale)
scaler.fit(unscaled_inputs)
scaled_inputs = scaler.transform(unscaled_inputs)
scaled_inputs
```

### inputs dataframe to nd.array

```python
scaled_inputs = scaled_inputs.to_numpy()
scaled_inputs
```

### shuffle the data

```python
scaled_inputs.shape[0]
```

```python
shuffled_indices = np.arange(scaled_inputs.shape[0])
np.random.shuffle(shuffled_indices)

shuffled_targets = targets[shuffled_indices]
shuffled_inputs = scaled_inputs[shuffled_indices]
```

# split the data into train, test and validation data

```python
sample_count = scaled_inputs.shape[0]

train_sample_count = int(0.8*sample_count)
test_sample_count = int(0.1*sample_count)
validation_sample_count = sample_count - train_sample_count - test_sample_count # not necessarily useful

train_inputs = shuffled_inputs[:train_sample_count]
train_targets = shuffled_targets[:train_sample_count]

test_inputs = shuffled_inputs[train_sample_count:train_sample_count+test_sample_count]
test_targets = shuffled_targets[train_sample_count:train_sample_count+test_sample_count]

validation_inputs = shuffled_inputs[train_sample_count+test_sample_count:]
validation_targets = shuffled_targets[train_sample_count+test_sample_count:]

print(f'Train difference: {train_inputs.shape[0]/np.sum(train_targets)}\n')
print(f'Validation difference: {validation_inputs.shape[0]/np.sum(validation_targets)}\n')
print(f'Test difference: {test_inputs.shape[0]/np.sum(test_targets)}\n')
```

```python
np.savez('absenteeism_data_train',inputs=train_inputs,targets=train_targets)
np.savez('absenteeism_data_validation',inputs=validation_inputs,targets=validation_targets)
np.savez('absenteeism_data_test',inputs=test_inputs,targets=test_targets)
```
